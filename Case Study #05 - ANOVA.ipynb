# 3.Checking whether all the zones generate the same amount of sales.

# We will use one-way ANOVA 

Null Hypothesis,H0 : All zones generate same amount of sales
    
Alternate Hypothesis,H1 : Sales of atleast one zone is different

```python
data.head()
```

```python
data.shape
```

```python
alpha=0.05
```

```python
#numerator degrees of freedom (k-1) = (4-1) = 3
dfn=3
#denominator degrees of freedom (N-k)= (29*4-4) = 112

dfd=112
```

```python
#critical f- value
f_critical=stats.f.ppf(q=.95,dfn=3,dfd=112)
print('Critical F - value = ',f_critical)
```

```python
f_value,p_value=stats.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
```

```python
print('Calculated F - value = ',f_value)

print('P - value = ',p_value)
```

# Another method using  STATSMODELS

```python

import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df_melt = pd.melt(data.reset_index(), id_vars=['index'], value_vars=['Zone - A', 'Zone - B', 'Zone - C', 'Zone - D'])
df_melt
```

```python
model = ols('value ~ C(variable)', data=df_melt).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

# Interpretation:

The F test statistic is 5.6720 and the corresponding p-value is 0.001183. Since the p-value is  less than .05 and Caculated F-value is greater than critical F-value we are forced to reject the null hypothesis (H0).

We can infer that mean of atleast one column is different ,ie, the sales varies with respect to the zones
